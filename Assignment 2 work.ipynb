```python
#Show a breakdown of distance from home by job role and attrition
```

```python
HRattrition1 = HRattrition[['Attrition', 'JobRole', 'DistanceFromHome', 'MonthlyIncome', 'Education' ]]
HRattrition1
```

```python
HRattrition1.groupby(['DistanceFromHome', 'JobRole']).count().reset_index()
```

```python
Attrited_yes = HRattrition1 [['DistanceFromHome', 'JobRole','Attrition' ]][HRattrition1['Attrition'] == 'Yes']
Attrited_yes.groupby(['DistanceFromHome','JobRole']).count().head(10)
```

```python
Attrited_no = HRattrition1[['DistanceFromHome', 'JobRole','Attrition' ]][HRattrition1['Attrition'] == 'No']
Attrited_no.groupby(['DistanceFromHome', 'JobRole',]).count().head(10)
```
